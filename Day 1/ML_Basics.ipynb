```python
import pandas as pd
from sklearn.linear_model import LinearRegression
data=pd.DataFrame({
    "sqft": [1000, 1200, 1500, 1800],
    "price": [5000000, 6000000, 7500000, 9000000]
})
X = data[["sqft"]]   # input
y = data["price"]
model=LinearRegression()
model.fit(X,y)
z=model.predict([[750]])
print(f"The output is {z}")
```

```python
# Step 1: Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Step 2: Load dataset
df = pd.read_csv("housing.csv")

# Step 3: See dataset structure
print("First 5 rows:")
print(df.head())

print("\nColumn names:")
print(df.columns)

# Step 4: Select input and output
X = df[['median_income']]          # input feature
y = df['median_house_value']       # output (house price)

# Step 5: Split data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Step 6: Create Linear Regression model
model = LinearRegression()

# Step 7: Train the model
model.fit(X_train, y_train)

# Step 8: Predict prices
predictions = model.predict(X_test)

# Step 9: Print some predictions
print("\nPredicted house prices:")
print(predictions[:10])
```
