```python
import sqlite3

conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Drop tables if they exist (so no error when re-running)
cursor.execute("DROP TABLE IF EXISTS employees")
cursor.execute("DROP TABLE IF EXISTS departments")

# Create tables
cursor.execute("""
CREATE TABLE departments (
    id INTEGER PRIMARY KEY,
    dept_name TEXT
)
""")

cursor.execute("""
CREATE TABLE employees (
    id INTEGER PRIMARY KEY,
    name TEXT,
    department_id INTEGER
)
""")

# Insert data
cursor.execute("INSERT INTO departments VALUES (1, 'HR')")
cursor.execute("INSERT INTO departments VALUES (2, 'Engineering')")
cursor.execute("INSERT INTO departments VALUES (3, 'Sales')")

cursor.execute("INSERT INTO employees VALUES (1, 'Riya', 2)")
cursor.execute("INSERT INTO employees VALUES (2, 'Aman', 1)")
cursor.execute("INSERT INTO employees VALUES (3, 'Neha', 2)")
cursor.execute("INSERT INTO employees VALUES (4, 'Raj', 4)")  # No matching department

conn.commit()
```

```python
#INNER JOIN
#Shows only matching rows in both tables.
cursor.execute("""
SELECT employees.name, departments.dept_name
FROM employees
INNER JOIN departments
ON employees.department_id = departments.id
""")

print("INNER JOIN:")
for row in cursor.fetchall():
    print(row)
```

```python
#LEFT JOIN
#ALL LEFT MENTIONED COLUMS AND MATCHING RIGHT OTHERWISE NONE
cursor.execute("""
SELECT employees.name, departments.dept_name
FROM employees
LEFT JOIN departments
ON employees.department_id = departments.id
""")

print("\nLEFT JOIN:")
for row in cursor.fetchall():
    print(row)
```

```python
#RIGHT JOIN
#ALL RIGHT MENTIONED COLUMS AND MATCHING LEFT OTHERWISE NONE
#SQLITE DOESNOT SUPPORT RIGHT JOIN IT SUPPORTS LEFT JOIN REVERSE
cursor.execute("""
SELECT employees.name, departments.dept_name
FROM employees
LEFT JOIN departments
ON employees.department_id = departments.id
""")

print("\nLEFT JOIN:")
for row in cursor.fetchall():
    print(row)
```

```python
#FULL JOINS
#SQL DOESN'T SUPPORT FULL JOIN SO LEFT AND RIGHT JOIN IN UNION OPERATOR
cursor.execute("""
SELECT employees.name, departments.dept_name
FROM employees
LEFT JOIN departments
ON employees.department_id = departments.id

UNION

SELECT employees.name, departments.dept_name
FROM departments
LEFT JOIN employees
ON employees.department_id = departments.id
""")

print("\nFULL JOIN (Simulated):")
for row in cursor.fetchall():
    print(row)
```

```python
#AGGREGATION FUNCTIONS
#An aggregation function in SQL is a function that performs a calculation on multiple rows and returns a single summarized value.

import sqlite3

# Connect to database
conn = sqlite3.connect("school.db")
cursor = conn.cursor()

# Drop table if exists (so it runs again without error)
cursor.execute("DROP TABLE IF EXISTS students")

# Create table
cursor.execute("""
CREATE TABLE students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    marks INTEGER
)
""")

# Insert data
cursor.execute("INSERT INTO students VALUES (1, 'Riya', 80)")
cursor.execute("INSERT INTO students VALUES (2, 'Aman', 90)")
cursor.execute("INSERT INTO students VALUES (3, 'Neha', 70)")
cursor.execute("INSERT INTO students VALUES (4, 'Raj', 85)")

conn.commit()

#SUM
cursor.execute("SELECT SUM(marks) FROM STUDENTS")
print("Total marks:", cursor.fetchone()[0])


#COUNT
cursor.execute("SELECT count(id) from students")
print("Total students",cursor.fetchone()[0])

#AVERAGE
cursor.execute("SELECT AVG(marks) FROM STUDENTS")
print("AVERAGE marks:", cursor.fetchone()[0])

#MAX
cursor.execute("SELECT MAX(marks) FROM STUDENTS")
print("MAXIMUM marks:", cursor.fetchone()[0])


#MIN
cursor.execute("SELECT MIN(marks) FROM STUDENTS")
print("MINIMUM marks:", cursor.fetchone()[0])
```

```python
#SUBQUERY IN SQL
#IT IS THE




#Find students who scored more than average marks.
import sqlite3

# Connect to database
conn = sqlite3.connect("school.db")
cursor = conn.cursor()

# Drop table if exists (so it runs without error)
cursor.execute("DROP TABLE IF EXISTS students")

# Create table
cursor.execute("""
CREATE TABLE students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    marks INTEGER
)
""")

# Insert data
cursor.execute("INSERT INTO students VALUES (1, 'Riya', 80)")
cursor.execute("INSERT INTO students VALUES (2, 'Aman', 90)")
cursor.execute("INSERT INTO students VALUES (3, 'Neha', 70)")
cursor.execute("INSERT INTO students VALUES (4, 'Raj', 85)")

conn.commit()

# Subquery Example:
# Find students who scored more than average marks
# We need result of one query
# To use that result inside another query
# To filter data based on another table
# When JOIN is not required
# To solve complex problems in simple steps


cursor.execute("""
SELECT name
FROM students
WHERE marks > (
    SELECT AVG(marks)
    FROM students
)
""")

print("Students scoring above average:")
for row in cursor.fetchall():
    print(row[0])

conn.close()
```

```python
#PRACTICE ASSIGNMENTS

# Task 1: The Customer-Order Report You are given two tables: Customers (CustomerID, CustomerName) and Orders (OrderID, CustomerID, OrderDate, Amount). Write a query to list all customers and their total spend. (Note: Some customers might have 0 orders—don't lose them!) Filter that list to show only customers who have spent more than $500 in total.
# import sqlite3

conn = sqlite3.connect("shop.db")
cursor = conn.cursor()

# Drop tables if exist
cursor.execute("DROP TABLE IF EXISTS Orders")
cursor.execute("DROP TABLE IF EXISTS Customers")

# Create Customers table
cursor.execute("""
CREATE TABLE Customers (
    CustomerID INTEGER PRIMARY KEY,
    CustomerName TEXT
)
""")

# Create Orders table
cursor.execute("""
CREATE TABLE Orders (
    OrderID INTEGER PRIMARY KEY,
    CustomerID INTEGER,
    OrderDate TEXT,
    Amount REAL
)
""")

# Insert Customers
cursor.execute("INSERT INTO Customers VALUES (1, 'Riya')")
cursor.execute("INSERT INTO Customers VALUES (2, 'Aman')")
cursor.execute("INSERT INTO Customers VALUES (3, 'Neha')")
cursor.execute("INSERT INTO Customers VALUES (4, 'Raj')")  # No orders

# Insert Orders
cursor.execute("INSERT INTO Orders VALUES (101, 1, '2024-01-01', 300)")
cursor.execute("INSERT INTO Orders VALUES (102, 1, '2024-02-01', 250)")
cursor.execute("INSERT INTO Orders VALUES (103, 2, '2024-03-01', 200)")
cursor.execute("INSERT INTO Orders VALUES (104, 3, '2024-04-01', 600)")

conn.commit()
```

```python
cursor.execute("""
SELECT Customers.CustomerName,
       IFNULL(SUM(Orders.Amount), 0) AS TotalSpend
FROM Customers
LEFT JOIN Orders
ON Customers.CustomerID = Orders.CustomerID
GROUP BY Customers.CustomerID
""")

print("All Customers with Total Spend:")
for row in cursor.fetchall():
    print(row)
```

```python
#TASK 3 Task 3: The Nested Logic Find the names of employees who work in the 'Engineering' department using a Subquery instead of a Join.

import sqlite3

conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Drop tables if exist
cursor.execute("DROP TABLE IF EXISTS employees")
cursor.execute("DROP TABLE IF EXISTS departments")

# Create departments table
cursor.execute("""
CREATE TABLE departments (
    id INTEGER PRIMARY KEY,
    dept_name TEXT
)
""")

# Create employees table
cursor.execute("""
CREATE TABLE employees (
    id INTEGER PRIMARY KEY,
    name TEXT,
    department_id INTEGER
)
""")

# Insert departments
cursor.execute("INSERT INTO departments VALUES (1, 'HR')")
cursor.execute("INSERT INTO departments VALUES (2, 'Engineering')")
cursor.execute("INSERT INTO departments VALUES (3, 'Sales')")

# Insert employees
cursor.execute("INSERT INTO employees VALUES (1, 'Riya', 2)")
cursor.execute("INSERT INTO employees VALUES (2, 'Aman', 1)")
cursor.execute("INSERT INTO employees VALUES (3, 'Neha', 2)")
cursor.execute("INSERT INTO employees VALUES (4, 'Raj', 3)")

conn.commit()

cursor.execute("""
SELECT name
FROM employees
WHERE department_id = (
    SELECT id
    FROM departments
    WHERE dept_name = 'Engineering'
)
""")

print("Employees in Engineering Department:")
for row in cursor.fetchall():
    print(row[0])

conn.close()
```

```python
# Task 2: The Performance Tracker (Aggregation) Goal: Find the average score for each candidate who has submitted work. Action: Join Candidates and Scores. Function: Use AVG(Points) and GROUP BY the candidate's name. Challenge: Only show candidates whose average score is above 80 (Hint: Use HAVING).

import sqlite3

conn = sqlite3.connect("performance.db")
cursor = conn.cursor()

# Drop tables if exist
cursor.execute("DROP TABLE IF EXISTS Scores")
cursor.execute("DROP TABLE IF EXISTS Candidates")

# Create Candidates table
cursor.execute("""
CREATE TABLE Candidates (
    CandidateID INTEGER PRIMARY KEY,
    CandidateName TEXT
)
""")

# Create Scores table
cursor.execute("""
CREATE TABLE Scores (
    ScoreID INTEGER PRIMARY KEY,
    CandidateID INTEGER,
    Points INTEGER
)
""")

# Insert Candidates
cursor.execute("INSERT INTO Candidates VALUES (1, 'Riya')")
cursor.execute("INSERT INTO Candidates VALUES (2, 'Aman')")
cursor.execute("INSERT INTO Candidates VALUES (3, 'Neha')")
cursor.execute("INSERT INTO Candidates VALUES (4, 'Raj')")  # No scores

# Insert Scores
cursor.execute("INSERT INTO Scores VALUES (101, 1, 85)")
cursor.execute("INSERT INTO Scores VALUES (102, 1, 90)")
cursor.execute("INSERT INTO Scores VALUES (103, 2, 70)")
cursor.execute("INSERT INTO Scores VALUES (104, 2, 75)")
cursor.execute("INSERT INTO Scores VALUES (105, 3, 95)")
cursor.execute("INSERT INTO Scores VALUES (106, 3, 88)")

conn.commit()


cursor.execute("""
SELECT Candidates.CandidateName,
       AVG(Scores.Points) AS AverageScore
FROM Candidates
JOIN Scores
ON Candidates.CandidateID = Scores.CandidateID
GROUP BY Candidates.CandidateName
HAVING AverageScore > 80
""")

print("Candidates with Average Score > 80:")
for row in cursor.fetchall():
    print(row)

conn.close()
```

```python
#Task 1: The Hardware Audit (Joins) Goal: Identify which candidates are currently working without a laptop. Action: Write a query using a LEFT JOIN between Candidates and Laptops. Filter: Use a WHERE clause to find only those where the SerialNo is NULL.


import sqlite3

conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Drop tables if exist
cursor.execute("DROP TABLE IF EXISTS Laptops")
cursor.execute("DROP TABLE IF EXISTS Candidates")

# Create Candidates table
cursor.execute("""
CREATE TABLE Candidates (
    CandidateID INTEGER PRIMARY KEY,
    CandidateName TEXT
)
""")

# Create Laptops table
cursor.execute("""
CREATE TABLE Laptops (
    LaptopID INTEGER PRIMARY KEY,
    CandidateID INTEGER,
    SerialNo TEXT
)
""")

# Insert Candidates
cursor.execute("INSERT INTO Candidates VALUES (1, 'Riya')")
cursor.execute("INSERT INTO Candidates VALUES (2, 'Aman')")
cursor.execute("INSERT INTO Candidates VALUES (3, 'Neha')")
cursor.execute("INSERT INTO Candidates VALUES (4, 'Raj')")

# Insert Laptops (Raj has no laptop)
cursor.execute("INSERT INTO Laptops VALUES (101, 1, 'SN123')")
cursor.execute("INSERT INTO Laptops VALUES (102, 2, 'SN456')")
cursor.execute("INSERT INTO Laptops VALUES (103, 3, 'SN789')")

conn.commit()


cursor.execute("""
SELECT Candidates.CandidateName
FROM Candidates
LEFT JOIN Laptops
ON Candidates.CandidateID = Laptops.CandidateID
WHERE Laptops.SerialNo IS NULL
""")

print("Candidates without Laptop:")
for row in cursor.fetchall():
    print(row[0])

conn.close()
```

```python
%%writefile library.py

def check_out_book(book_list, book_name):
    if book_name in book_list:
        book_list.remove(book_name)
        return book_name
    else:
        raise KeyError(f"{book_name} not found in library")

```

```python
%%writefile test_library.py

import pytest
from library import check_out_book

# Happy Path Test
def test_check_out_book_success():
    books = ["Python", "DBMS", "AI"]
    result = check_out_book(books, "Python")
    assert result == "Python"
    assert "Python" not in books


# Edge Case Test
def test_check_out_book_not_found():
    books = ["Python", "DBMS", "AI"]
    with pytest.raises(KeyError):
        check_out_book(books, "ML")
```
