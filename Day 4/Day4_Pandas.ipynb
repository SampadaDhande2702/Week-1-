```python
import pandas as pd
import numpy as np

# 1. SERIES & DATAFRAME CREATION
# Creating a series for product IDs and a DataFrame for sales
products = pd.Series(['ball', 'pen', 'pencil', 'mug', 'ashtray'], name='item')
dates = pd.date_range('20230101', periods=5)
sales_df = pd.DataFrame({
    'date': dates,
    'item': ['ball', 'pen', 'pencil', 'ball', 'mug'],
    'units': [10, 15, 12, 10, 8] # ball is duplicated for demonstration
}, index=[0, 1, 2, 3, 4])

# 2. DATA TRANSFORMATION: Removing Duplicates & Mapping
# Detecting and removing duplicates
sales_df = sales_df.drop_duplicates(subset=['item', 'units'])

# Adding values via mapping (Mapping product names to prices)
prices = {'ball': 5.5, 'pen': 1.3, 'pencil': 0.6, 'mug': 4.2, 'ashtray': 2.7}
sales_df['price_per_unit'] = sales_df['item'].map(prices)

# 3. DATA PREPARATION: Merging & Removing
# Creating another DataFrame for inventory
inv_df = pd.DataFrame({
    'item': ['ball', 'pen', 'pencil', 'mug'],
    'stock': [100, 200, 150, 80]
})

# Merging DataFrames (Inner Join)
final_df = pd.merge(sales_df, inv_df, on='item')

# Removing an unwanted column
# del final_df['units'] # Alternative: final_df.drop()

# 4. INDEXING & SELECTING DATA
# Using .loc for label-based selection
high_sales = final_df.loc[final_df['price_per_unit'] > 1.0]

# Using .iloc for position-based selection (first 2 rows)
top_rows = final_df.iloc[:2]

# 5. PIVOTING & RESHAPING
# Long to Wide format
wide_df = final_df.pivot(index='date', columns='item', values='price_per_unit')

# Stacking/Unstacking
stacked = final_df.set_index(['date', 'item']).stack()

# 6. DISCRETIZATION & BINNING
# Binning prices into categories
bins = [0, 1, 5, 10]
labels = ['Cheap', 'Mid', 'Expensive']
final_df['price_cat'] = pd.cut(final_df['price_per_unit'], bins, labels=labels)

# 7. STATISTICAL ANALYSIS & AGGREGATION
# Summary statistics
print("Stats:\n", final_df.describe())

# Data Aggregation using GroupBy
total_revenue = (final_df['units'] * final_df['price_per_unit']).sum()
group_avg = final_df.groupby('price_cat')['units'].mean()

# 8. DEALING WITH DATE & TIME
# Creating a sequence of dates and a Timedelta
final_df['delivery_date'] = final_df['date'] + pd.Timedelta(days=3)

print("\nFinal Processed DataFrame:")
print(final_df)
```
