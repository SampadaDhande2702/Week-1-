```python
# Mutable and Immutable Data Types in Python

# What is Mutable?
# A mutable object is one whose value can be changed after creation.

numbers = [1, 2, 3]
numbers[0] = 10
print(numbers)

# 3. What is Immutable?
# An immutable object is one whose value cannot be changed after creation.

name = "Sampada"
name[0] = "R"


```

```python
#Loops in python

#1)While loop
i=1
while i<=5:
  print(i)
  i=i+1

```

```python
#For loop
password = "AI2026"

for letter in password:
    print("Scanning letter:", letter)

print("Scan Complete ")
```

```python
#Break
#We do use break to break the iteration of loops which runs continiously until any specific condition met

password = "AI2026"

for letter in password:
    print("Scanning letter:", letter)
    if letter=="2":
      break

print("Scan Complete ")
```

```python
#Continue
#Use to pass that particular part
password = "AI2026"

for letter in password:

    if letter=="2":
      continue
    print("Scanning letter:", letter)

print("Scan Complete ")
```

```python
#Functions in python

#In Python, a function is a block of organized and reusable code that performs a specific task.

#Syntax of a Function
#   def function_name():
       # block of code

def greet():
    print("Good morning")

greet()
```

```python
# 4 Types of functions

# Function with no arguments and no return value
# Function with arguments and no return value
# Function with no arguments but with return value
# Function with arguments and with return value

# 1) Function with No Arguments and No Return Value

def greet():
    print("Good morning")

greet()

# 2) 2. Function with Arguments and No Return Value


def name(name):
  print("Hello",name)

name("Sampada")


# 3) Function with No Arguments but With Return Value

def get_number():
    return 10

num = get_number()
print(num)


# 4) 4. Function with Arguments and With Return Value

def add(a, b):
    return a + b

result = add(5, 3)
print(result)


```

```python
#Types of Function Arguments

#1. Default Arguments

#A default argument is a parameter that assumes a default value if a value is not provided in the function call for that argument.

def myFun(x, y=50):
    print("x: ", x)
    print("y: ", y)

myFun(10)

# 2. Keyword Arguments

def student(fname, lname):
    print(fname, lname)

student(fname='Geeks', lname='Practice')
student(lname='Practice', fname='Geeks')


# 3. 3. Positional Arguments

def nameAge(name, age):
    print("Hi, I am", name)
    print("My age is ", age)

print("Case-1:")
nameAge("Suraj", 27)

print("\nCase-2:")
nameAge(27, "Suraj")
```

```python
#Lambda function

add = lambda a, b: a + b
print(add(5, 3))
```
